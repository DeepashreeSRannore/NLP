```python
import spacy

# Load spaCy blank English model
lang_model = spacy.blank("en_core_web_sm")
email_ids = []

file_path = r'C:\\Users\\EmailExtractionDataset.txt'

# Process each line separately
with open(file_path, 'r', encoding='utf-8') as file:
    for line in file:
        doc = lang_model(line.strip())
        email_ids.extend([token.text for token in doc if token.like_email])

print(f"Extracted {len(email_ids)} email(s).")
print(email_ids)
```
